# Optimize y = (x-1)^2


```python
from ax.api.client import Client
from ax.api.configs import RangeParameterConfig
```

```python
def cal_y(x):
    return (x - 1) ** 2
```

```python
# 1. Initialize the Client.
client = Client()
```

```python
# 2. Configure where Ax will search.
client.configure_experiment(
    name="quadratic_optimization",
    parameters=[
        RangeParameterConfig(
            name="x",
            bounds=(-10.0, 10.0),
            parameter_type="float",
        ),
    ],
)
```

```python
client.configure_optimization(objective="-y")
```

```python
for i in range(20):
    data = client.get_next_trials(max_trials=1)
    print(data)

    for trial_index, parameters in data.items():
        print(f"Trial {trial_index} with parameters {parameters}")
        client.complete_trial(
            trial_index=trial_index,
            raw_data={"y": cal_y(parameters.get("x", 0))},
        )
```
